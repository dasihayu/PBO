**Praktikum 1 : Tipe Data Dasar**

```python
# 1. Number (int, float, complex)

# Integer (int)
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))

# Float
angka_float = 3.14
print("angka_float =", angka_float, "->", type(angka_float))

# Complex
angka_complex = 2 + 3j
print("angka_complex =", angka_complex, "->", type(angka_complex))

# 2. Boolean
is_active = True
print("is_active =", is_active, "->", type(is_active))

# 3. String
teks = "Hello, Python!"
print("teks =", teks, "->", type(teks))

# 4. List
daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka =", daftar_angka, "->", type(daftar_angka))

# 5. Tuple
koordinat = (10, 20)
print("koordinat =", koordinat, "->", type(koordinat))

# 6. Dictionary
data_mahasiswa = {
    "nama": "Andi",
    "nim": "A11.2022.12345",
    "jurusan": "Teknik Informatika"
}
print("data_mahasiswa =", data_mahasiswa, "->", type(data_mahasiswa))

# 7. Set
himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka =", himpunan_angka, "->", type(himpunan_angka))

# 8. Konversi tipe data
nilai_str = "100"
print("\nnilai_str =", nilai_str, "->", type(nilai_str))

nilai_int = int(nilai_str)
print("nilai_int =", nilai_int, "->", type(nilai_int))
```

**Praktium 2 : Variable dan operasi dasar**

```python
# -*- coding: utf-8 -*-

# 1. Variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmetika
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b
pembagian_bulat = a // b
modulus = a % b
pangkat = a ** b

print("\nOPERASI ARITMETIKA")
print("Penjumlahan =", penjumlahan)
print("Pengurangan =", pengurangan)
print("Perkalian =", perkalian)
print("Pembagian =", pembagian)
print("Pembagian Bulat =", pembagian_bulat)
print("Modulus =", modulus)
print("Pangkat =", pangkat)

# 3. Perbandingan
print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a < b =", a < b)
print("a == b =", a == b)
print("a != b =", a != b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

# 4. Logika
x = True
y = False

print("\nOPERASI LOGIKA")
print("x and y =", x and y)
print("x or y =", x or y)
print("not x =", not x)

# 5. Percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi")
else:
    print("\nKondisi tidak terpenuhi")
```

**Praktikum 3 : Percabangan**

```python
# IF sederhana
nilai = 85
if nilai > 80:
    print("Lulus dengan nilai tinggi")

# IF-ELSE
umur = 17
if umur >= 18:
    print("Boleh bikin SIM")
else:
    print("Belum cukup umur")

# IF-ELIF-ELSE
hari = "Rabu"
if hari == "Senin":
    print("Kerja")
elif hari == "Selasa":
    print("Rapat")
elif hari == "Rabu":
    print("Diskon")
else:
    print("Hari biasa")

# Nested IF
suhu = 35
if suhu > 30:
    print("Panas")
    if suhu > 40:
        print("Sangat panas")
    else:
        print("Masih normal")
else:
    print("Sejuk")

# Logika
nilai_teori = 75
nilai_praktik = 80

if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Lulus")
elif nilai_teori < 70:
    print("Perbaiki teori")
else:
    print("Perbaiki praktik")

# Ternary
angka = -5
status = "Positif" if angka > 0 else "Negatif/Nol"
print(status)
```

**Praktikum 4 - Perulangan**

```python
# FOR
for i in range(5):
    print(i)

# List
buah = ["apel", "mangga", "jeruk"]
for b in buah:
    print(b)

# WHILE
count = 0
while count < 5:
    print(count)
    count += 1

# BREAK
for i in range(10):
    if i == 3:
        break
    print(i)

# CONTINUE
for i in range(5):
    if i == 2:
        continue
    print(i)

# Nested Loop
for i in range(3):
    for j in range(2):
        print(i, j)

# ELSE Loop
for x in range(3):
    print(x)
else:
    print("Selesai")

# PASS
for i in range(3):
    if i == 1:
        pass
    print(i)
```

**Penugasan 1 : BMI**

```python
berat = float(input("Masukkan berat badan (kg): "))
tinggi_cm = float(input("Masukkan tinggi badan (cm): "))

tinggi_m = tinggi_cm / 100
bmi = berat / (tinggi_m ** 2)

print("BMI =", round(bmi, 2))

if bmi < 18.5:
    print("Kategori: Underweight")
elif bmi < 25:
    print("Kategori: Normal")
elif bmi < 30:
    print("Kategori: Overweight")
else:
    print("Kategori: Obesitas")
```

**Program 2 — Genap, Ganjil, Prima**

```python
angka = int(input("Masukkan angka: "))

# Genap / Ganjil
if angka % 2 == 0:
    print("Bilangan Genap")
else:
    print("Bilangan Ganjil")

# Prima
if angka > 1:
    prima = True
    for i in range(2, angka):
        if angka % i == 0:
            prima = False
            break

    if prima:
        print("Bilangan Prima")
    else:
        print("Bukan Prima")
else:
    print("Bukan Prima")
```
