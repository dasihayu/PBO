**Praktikum 01 — Fungsi pada Python**

```python
# Fungsi Built-in
print("Ini adalah contoh fungsi built-in")

kata = "Pemrograman"
panjang_kata = len(kata)
print(f"Panjang kata '{kata}' adalah: {panjang_kata}")

angka = [10, 5, 30, 40, 25]
nilai_max = max(angka)
print(f"Nilai maksimum dalam daftar {angka} adalah: {nilai_max}")

# Fungsi User-Defined
def cetak_kuadrat(angka):
    print(f"Kuadrat dari {angka} adalah: {angka ** 2}")

def hitung_luas_persegi_panjang(panjang, lebar):
    return panjang * lebar

def info_mahasiswa(nama, umur, ipk):
    print(f"Nama: {nama}, Umur: {umur}, IPK: {ipk}")

def sapa_pengguna(nama):
    print(f"Halo, {nama}! Selamat datang di dunia Python.")

def hitung_keliling_persegi(sisi):
    return 4 * sisi

# Pemanggilan fungsi
cetak_kuadrat(5)

luas = hitung_luas_persegi_panjang(10, 5)
print(f"Luas persegi panjang: {luas}")

info_mahasiswa("Budi", 22, 3.8)

sapa_pengguna("Andi")

keliling = hitung_keliling_persegi(5)
print(f"Keliling persegi: {keliling}")
```

**Praktikum 03 — Kelas Sederhana**

```python
class Buku:
    def __init__(self, judul, pengarang, tahun):
        self.judul = judul
        self.pengarang = pengarang
        self.tahun = tahun
        self.status = "Tersedia"

    def tampilkan_info(self):
        print(self.judul, self.pengarang, self.tahun, self.status)

    def pinjam(self):
        if self.status == "Tersedia":
            self.status = "Dipinjam"
            print("Dipinjam")
        else:
            print("Sudah dipinjam")

    def kembalikan(self):
        self.status = "Tersedia"


buku1 = Buku("Python", "John", 2021)
buku1.tampilkan_info()
buku1.pinjam()
buku1.tampilkan_info()

class Mahasiswa:
    def __init__(self, nama, nim, umur):
        self.nama = nama
        self.nim = nim
        self.umur = umur

    def tampilkan_info(self):
        print(self.nama, self.nim, self.umur)

    def ubah_umur(self, umur_baru):
        self.umur = umur_baru


m1 = Mahasiswa("Andi", "123", 20)
m1.tampilkan_info()
m1.ubah_umur(21)
m1.tampilkan_info()
```

**Praktikum 04 — Method dalam Kelas**

```python
class Mobil:
    def __init__(self, merk, warna, tahun, harga):
        self.merk = merk
        self.warna = warna
        self.tahun = tahun
        self.harga = harga

    def tampilkan_info(self):
        print(self.merk, self.warna, self.tahun, self.harga)

    def diskon(self, persen):
        harga_baru = self.harga - (self.harga * persen / 100)
        print("Harga setelah diskon:", harga_baru)

    def hitung_usia(self, tahun_sekarang):
        return tahun_sekarang - self.tahun

    def update(self, harga_baru, tahun_baru):
        self.harga = harga_baru
        self.tahun = tahun_baru


mobil1 = Mobil("Toyota", "Hitam", 2015, 300000000)
mobil1.tampilkan_info()
mobil1.diskon(10)

print("Usia:", mobil1.hitung_usia(2025))

mobil1.update(280000000, 2022)
mobil1.tampilkan_info()
```

**Sistem Penjualan HP**

```python
%%writefile product.py

class HP:
    def __init__(self, nama, harga, stok):
        self.nama = nama
        self.harga = harga
        self.stok = stok

    def tampilkan(self):
        print(f"{self.nama} - Rp{self.harga} (Stok: {self.stok})")

    def beli(self, jumlah):
        if jumlah <= self.stok:
            self.stok -= jumlah
            total = self.harga * jumlah
            print("Total bayar:", total)
        else:
            print("Stok tidak cukup")
```

```python
from product import HP

def main():
    hp1 = HP("iPhone 13", 15000000, 5)
    hp2 = HP("Samsung S23", 12000000, 3)

    hp1.tampilkan()
    hp2.tampilkan()

    hp1.beli(2)
    hp1.tampilkan()

if __name__ == "__main__":
    main()
```
